```python
def get_data(numbers, ownership):
    keys = numbers.get("financials").get('quarterly').get('Fiscal Year')
    keyReleaseDates = [monthsChange(keys[i]) for i in range(len(keys))]
    ownDF = pd.DataFrame(ownership.get("insti_owner"),columns=["data","percent","shares"])

    ownDF['data'] = ownDF["data"].apply(lambda x: "-".join(x.split("-")[0:2]))
    hash = listToHashmap(keyReleaseDates)
    backHash = {v:k for k,v in hash.items()}

    mappings = []
    dates=[]
    for i in range(len(ownDF['data'])):
        date = ownDF['data'].iloc[i]
        check = hash.get(date)
    
        while(check==None):
            date = monthsChange(date,-1)
            check = hash.get(date)

        mappings.append(check)
        dates.append(date)
    #add column mappings to ownDF
    ownDF['mappings'] = mappings
    ownDF['dates'] = dates


    idxlist = []
    hashmap = {}
    #swap to negative range through ownDF['mappings'] (iterate backwards)
    for i in range(len(ownDF['mappings'])-1,-1,-1):
        thing = ownDF['mappings'][i]
        if(hashmap.get(thing)==None):
            idxlist.append(i)
        hashmap[thing] = 1

    #reverse idxlist

    idxlist = idxlist[::-1]

    repDF = ownDF.iloc[idxlist]
    import numpy as np
    vals = repDF['shares'].values
    diffs = np.diff(repDF['shares'].values)
    #padd diffs with 1 at end
    diffs = np.append(diffs,1)

    percentChange = (diffs / vals)

    #all but the last element
    percentChange = percentChange[:-1]

    #remove the first row of repDF and set percentChange column
    labelDF = repDF.iloc[1:]
    labelDF['percentChange'] = percentChange

    tempdf = pd.DataFrame()

    metrics ="""per_share_data_array
common_size_ratios
income_statement
balance_sheet
cashflow_statement
valuation_ratios
valuation_and_quality""".split("\n")
    names = ["dates"]
    lists = [keyReleaseDates]
    for j in metrics: 
      
       
        for i in numbers.get("financials").get("quarterly").get(j).keys():
            names.append(j+"/"+i)
            lists.append(numbers.get("financials").get("quarterly").get(j).get(i))

    #concatenate all lists using the names as columns using .concat
    tempdf = pd.concat([tempdf,pd.DataFrame(lists).T],axis=1)
    tempdf.columns = names
    nonnumerical = ['common_size_ratios/RD-to-Revenue','valuation_and_quality/Restated Filing Date','valuation_and_quality/Filing Date','valuation_and_quality/Earnings Release Date']
    def fix(x):
        try:
            return float(x)
        except:
            return 0
    for i in nonnumerical:
        if(i=="dates"):
            continue
        try:
            tempdf = tempdf.drop(i,axis=1)
        except:
            print("Already removed " + i)

    #float every number in tempdf
    for i in tempdf.columns:
        if(i=="dates"):
            continue
        tempdf[i] = tempdf[i].apply(lambda x: fix(x))
    inQuestion = listToHashmap([backHash[i] for i in finDF['mappings']])
   
    newIdxList= []
    for i in range(len(tempdf['dates'])):
        thing = inQuestion.get(tempdf['dates'][i])
        if(thing!=None):
            newIdxList.append(i)
    dataDF  = tempdf.iloc[newIdxList]
    dataDF['labels'] = labelDF[["percentChange"]].values
    #reset indicies
    dataDF = dataDF.reset_index(drop=True)
 
    return dataDF


```
